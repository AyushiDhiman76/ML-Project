##Which actor has acted in the most number of movies? Deep dive into the movies, genres and pro ts corresponding to this actor

```python
actor_counts = {}

for cast_list in df['cast'].dropna():
    for actor in eval(cast_list):
        actor_counts[actor['name']] = actor_counts.get(actor['name'], 0) + 1

most_prolific_actor = max(actor_counts, key=actor_counts.get)
print(f"Most Prolific Actor: {most_prolific_actor}")
print(f"Number of Movies: {actor_counts[most_prolific_actor]}")
```

##Top 3 directors prefer which actors the most

```python
import pandas as pd

df = pd.read_csv("imdb_data.csv")

df['director'] = df['crew'].str.extract(r'{"job": "Director", "name": "(.*?)"}')
df['cast_list'] = df['cast'].apply(lambda x: [actor['name'] for actor in eval(x)] if pd.notna(x) else [])

director_actor_counts = {}

for index, row in df.iterrows():
    director = row['director']
    if director:
        for actor in row['cast_list']:
            director_actor_counts.setdefault(director, {}).setdefault(actor, 0)
            director_actor_counts[director][actor] += 1

top_directors = sorted(director_actor_counts, key=lambda x: len(director_actor_counts[x]), reverse=True)[:3]
for director in top_directors:
    preferred_actors = sorted(director_actor_counts[director].items(), key=lambda item: item[1], reverse=True)[:3]
    print(f"Director: {director}")
    for actor, count in preferred_actors:
        print(f"  - {actor}: {count} collaborations")
```
